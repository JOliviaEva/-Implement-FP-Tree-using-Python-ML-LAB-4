# ***Scenario 1: Movie Streaming Platform***

Create a class to represent each node in the FP-Tree

```python
from collections import Counter
class TreeNode:
    def __init__(self, name):
        self.name = name
        self.count = 0
        self.children = {}

    def display(self, ind=1):
        print('  ' * ind + f"└─ {self.name} : {self.count}")
        for child in self.children.values():
            child.display(ind + 1)
```

STEP 1: Dataset & Minimum Support

```python
dataset = [
    ['Action', 'Comedy', 'Thriller'],
    ['Action', 'Drama', 'Thriller'],
    ['Comedy', 'Thriller', 'Romance'],
    ['Action', 'Comedy', 'Thriller'],
    ['Action', 'Comedy', 'Drama']
]
min_support = 3
```

STEP 2 & 3: Count and Filter Frequent Items

```python
# Flatten list to count all items, then filter by minimum support
all_items = [item for transaction in dataset for item in transaction]
freq_items = {item: count for item, count in Counter(all_items).items() if count >= min_support}

print(f"Frequent Items (>= {min_support}): {freq_items}\n")
```

STEP 4 & 5: Process Transactions & Build FP-Tree

```python
root = TreeNode('Root')

for transaction in dataset:
    # Filter out infrequent items and sort remaining by frequency (descending)
    ordered_trans = sorted([item for item in transaction if item in freq_items],
                           key=lambda x: (-freq_items[x], x))
    if not ordered_trans:
        continue
    print(f"Ordered Transaction: {ordered_trans}")

    # Build the tree node-by-node
    current_node = root
    for item in ordered_trans:

        if item not in current_node.children:
            current_node.children[item] = TreeNode(item)

        current_node = current_node.children[item]
        current_node.count += 1

print("\n--- Final Constructed FP-Tree ---")
root.display()
```

# ***Scenario 2: Smart Home IoT System***

```python
from collections import Counter
class Node:
    def __init__(self, name):
        self.name = name
        self.count = 0
        self.children = {}

    def display(self, level=1):
        print('  ' * level + f"└─ {self.name} : {self.count}")
        for child in self.children.values():
            child.display(level + 1)
dataset = [
    ['Smart Light', 'Smart TV', 'Air Conditioner'],
    ['Smart Light', 'Smart Speaker', 'Air Conditioner'],
    ['Smart TV', 'Air Conditioner', 'Smart Speaker'],
    ['Smart Light', 'Smart TV', 'Air Conditioner'],
    ['Smart Light', 'Air Conditioner', 'Smart Speaker']
]
min_support = 3
device_counts = Counter(device for transaction in dataset for device in transaction)

# Filter items by min_support and sort them (Highest count first, alphabetical on tie)
frequent_devices = sorted(
    [item for item, count in device_counts.items() if count >= min_support],
    key=lambda x: (-device_counts[x], x)
)

#Step 1: Frequent Devices Master List
for device in frequent_devices:
    print(f"{device}: {device_counts[device]}")
print("\n")

root = Node("Root")

#Step 2: Processing Ordered Transactions
for i, transaction in enumerate(dataset, 1):
    ordered_trans = [device for device in frequent_devices if device in transaction]
    print(f"T{i}: {ordered_trans}")

    # Insert items into the tree
    current_node = root
    for device in ordered_trans:
        if device not in current_node.children:
            current_node.children[device] = Node(device)
        current_node = current_node.children[device]
        current_node.count += 1

print("\n--- Step 3: Final Constructed FP-Tree ---")
root.display()
```
